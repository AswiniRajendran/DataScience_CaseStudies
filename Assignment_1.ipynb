## Question-5: Meaningful visualisations(Atleast three)

 1. SCATTER PLOT

```python
plt.scatter(data['sepal length (cm)'],data['sepal width (cm)'])
plt.scatter(data['petal length (cm)'],data['petal width (cm)'],c='Red')
plt.show()
```

As of the linear relationship between Sepal length and width, they are visibly unrelated. But The petal length and width might have a chance for a positive correlation. 

2. HEATMAP

```python
sns.heatmap(data.corr(), annot=True,cmap='YlGnBu')
```

The highly related variables are sepal length with petal length and petal width, petal length and petal width. 

3. KDE PLOT

```python
sns.kdeplot(data['sepal length (cm)'], label='sepal length')
sns.kdeplot(data['sepal width (cm)'], label = 'sepal width' )
sns.kdeplot(data['petal length (cm)'],label = 'petal length')
sns.kdeplot(data['petal width (cm)'],label = 'petal width')
plt.xlabel('features')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

The distribution of each of these variables are plotted on the same graph. The variable, "Sepal Width" bell shaped, with an extreme amount of kurtosis. Sepal length is almost symmetric and the other two are not.

4. BOXPLOT

```python
sns.boxplot(data = [data['sepal length (cm)'],data['sepal width (cm)'],data['petal length (cm)'],data['petal width (cm)']], palette = ['Red','Blue','Green','Violet'])
plt.xlabel(['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)'])
plt.ylabel('values')
plt.show()
```

Box plot identifies outliers from sepal width, plots the median values of all four variables.
