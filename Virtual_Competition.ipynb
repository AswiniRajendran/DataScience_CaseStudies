```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("C:/Users/Ammu/Downloads/train_data.csv")
```

Exploratory Analysis

```python
train
```

```python
train = train.set_index('Loan_ID')
```

```python
train.info()
```

```python
numerical_features = [feature for feature in train.columns if train[feature].dtypes!='O']
categorical_features = [feature for feature in train.columns if train[feature].dtypes=='O']
```

```python
for i in numerical_features:
    if (train[i].isnull().sum() !=0):
        train[i]= train[i].fillna(train[i].median())
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
imputer = imputer.fit(train[categorical_features])
train[categorical_features] = imputer.transform(train[categorical_features])
train[categorical_features]
```

```python
train.info()
# result must show that there is no null values in the dataset.
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
train['Loan_Status']=LabelEncoder().fit_transform(train['Loan_Status'])
train['Loan_Status']
```

```python
# Gender, Married, Self_employed, Property_Area, Loan_Status, Education - Nominal - One-Hot Encoding
train = pd.get_dummies(train)
train
```

```python
xtrain=train.drop('Loan_Status',axis=1)
clms=xtrain.columns
ytrain=train['Loan_Status']
```

```python
from sklearn.preprocessing import StandardScaler
xtrain = StandardScaler().fit_transform(xtrain)
xtrain = pd.DataFrame(xtrain, columns=clms)
xtrain
```

```python
train['Loan_Status'].value_counts(normalize=True)
```

```python
xtest.info()
```

```python
numerical_features = [feature for feature in xtest.columns if xtest[feature].dtypes!='O']
categorical_features = [feature for feature in xtest.columns if xtest[feature].dtypes=='O']
```

```python
# Stratified Cross Validation
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score

# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
rf = rf.fit(xtrain, ytrain)

# function- skfold_validator
skfold_validator = StratifiedKFold(n_splits=10)
cv_result= cross_val_score(rf,xtrain,ytrain,cv=skfold_validator)
np.mean(cv_result)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
rf_random.fit(xtrain, ytrain)
```

```python
rf_random.best_params_
```

```python
def convert_array(array):
    converted_array = []
    for element in array:
        if element == 0:
            converted_array.append('N')
        elif element == 1:
            converted_array.append('Y')
        else:
            # Handle any other values if necessary
            converted_array.append(element)
    return converted_array
```

```python
yrf = convert_array(yrf)
print(yrf)
```
